###Import

```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Реестр_выполненных_рейсов_за_период (1).xlsx', header=1).drop(columns=['Unnamed: 0', 'Unnamed: 13', 'Unnamed: 15'])
df.dropna()
```

###Распределение значений признаков

```python
sns.regplot(x="Вместимость", y="Обслужено пассажиров", data=df, line_kws={"color": "yellow"});
```

```python
sns.scatterplot(x=df['МВМ, тонн'], y=df['Baggage Weight'], hue=df['МВМ, тонн']);
```

```python
x1 = df[df['Направление'] == 'International']['Кл.3']
x2 = df[df['Направление'] == 'Domestic']['Кл.3']
x = [x1, x2]

fig, ax = plt.subplots()
ax.hist(x)
labels = ["International", "Domestic"]
plt.legend(labels=labels)
plt.show();
```

###Boxplot по Bagagge weight

```python
fig = px.box(df, x="Baggage Weight")
fig.show()
```

###Piechart по В/П

```python
data = df['В/П'].value_counts(normalize=True)

fig, ax = plt.subplots()
ax.pie(data, labels=data.index, autopct='%1.1f%%');
```

###Pairplot

```python
sns.pairplot(df, hue="Направление");
```

###Sunburst plot

```python
df1 = df.dropna().copy()
df1[df1.isna().any(axis=1)]
fig = px.sunburst(df1, path=['В/П', 'Направление'])
fig.show()
```
