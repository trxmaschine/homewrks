```python
import random

from hypex import ABTest
from hypex.dataset import Dataset, InfoRole, TargetRole, TreatmentRole
```

```python
data = Dataset(
    roles={
        "flight_number": InfoRole(),
        "group": TreatmentRole(),
        "bag_dep_weight": TargetRole(),
        "pax_dep": TargetRole(),
        "pax_dep_economy": TargetRole()
         }, data=df,
)
data
```

```python
df = pd.read_sql("select * from traffic", engine)
df["group"] = [random.choice([0, 1]) for _ in range(len(df))]
df
```

```python
data.roles
```

```python
test = ABTest()
result = test.execute(data)
```

```python
result.resume
```

```python
import numpy as np
```

```python
new_data = data.data.copy()
new_data['bag_dep_weight'] = np.where(new_data['group'] == 1, new_data['bag_dep_weight'] * 1.3, new_data['bag_dep_weight'])
new_data
```

```python
data = Dataset(
    roles={
        "flight_number": InfoRole(),
        "group": TreatmentRole(),
        "bag_dep_weight": TargetRole(),
        "pax_dep": TargetRole(),
        "pax_dep_economy": TargetRole()
         }, data=new_data,
)
data
```

```python
test = ABTest()
result = test.execute(data)
result.resume
```

```python
test = ABTest(additional_tests=['t-test', 'u-test', 'chi2-test'])
result = test.execute(data)
result.resume
```

```python
test = ABTest(multitest_method="bonferroni")
result = test.execute(data)
result.resume
```
